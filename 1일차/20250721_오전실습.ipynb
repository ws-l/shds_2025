```python
adv = pd.read_csv("adv.csv", index_col=0)
adv.sales.mean()  #adv.mean() #adv.describe()
def func2(x):
  if x > 14.02:
    return "good"
  else:
    return "bad"
adv["sales2" ] = adv.sales.apply( func2 )
#adv.assign( sales2=adv.sales.apply( func2 )  )
adv.sales2.unique()
adv.sales2.value_counts()

#creditset2 ->credit
#default... 컬럼->value_counts
#income, loan, age-> describe
#new_col: loan > income*0.5 True,
credit = pd.read_csv("creditset2.csv")
credit.default10yr.value_counts()
credit[["income", "age", "loan"]].describe()
credit.drop("default10yr", axis=1).describe()
credit["new_col"] = credit.loan > credit.income * 1.2
credit.new_col.value_counts()
credit.drop( ["new_col"], axis=1, inplace=True)
credit.head()
```
