```python
#adv.csv 읽으셔서 sales가 y, tv가 X
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import seaborn as sns
df = pd.read_csv("adv.csv", index_col=[0])
#sns.pairplot( df ) #df.corr()
#파티셔닝, test에 대해서 예측값, MSE측정
X_train, X_test, y_train, y_test = train_test_split(df.drop("sales", axis=1),
          df["sales"], train_size=0.8, random_state=0)
model = RandomForestRegressor(n_estimators=50) #LinearRegression()
model.fit(X_train, y_train )
pred = model.predict( X_test )
((y_test - pred)**2).mean()  #MSE

#model.coef_    #직선식 기울기, 계수, coefficient
```

```python
#AMZN.csv -> y: Close(t), x1: Close(t-1), x2: Close(t-2), x3: Close(t-3)
df = pd.read_csv("AMZN.csv", index_col=[0])
df = df[["Close"]]
df["Close_t_1"] = df.Close.shift(1)
df["Close_t_2"] = df.Close.shift(2)
df["Close_t_3"] = df.Close.shift(3)

#파티셔닝, test에 대해서 예측값, MSE측정
X_train, X_test, y_train, y_test = train_test_split(df.drop("Close", axis=1),
          df["Close"], train_size=0.8, random_state=0)
```

```python
for i in [40,50,60]:
  model = RandomForestRegressor(n_estimators=i, random_state=0) #LinearRegression()
  model.fit(X_train, y_train )
  pred = model.predict( X_test )
  print( (((y_test - pred)**2).mean())**0.5  ) #RMSE
```

```python
from pykrx import stock
df = stock.get_market_ohlcv_by_date(ticker="005930", fromdate= "20240101", todate='20241231')
df = df[["종가"]]
df["종가_t_1"] = df.종가.shift(1)
df["종가_t_2"] = df.종가.shift(2)
df = df.dropna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop("종가", axis=1),
          df["종가"], train_size=0.8, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
#max_depth, n_estimators
params = { 'n_estimators':[50,100],
          'max_depth':[3,4]
          }
gs = GridSearchCV(estimator=RandomForestRegressor(),
             param_grid=params,
             cv = 5
)  #model
gs.fit(X_train, y_train)
gs.best_params_
gs.best_estimator_  #model
```

```python
model2 = gs.best_estimator_
model2.predict(X_test)
```

```python
df = pd.read_csv("BostonHousing.csv")
X = df.drop("medv", axis=1)
y = df.medv
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error

model1 = LinearRegression()
model1.fit(X_train, y_train)
print( mean_squared_error(y_test , model1.predict(X_test)) )

model2 = RandomForestRegressor()
model2.fit(X_train, y_train)
print( mean_squared_error(y_test , model2.predict(X_test)) )

model4 = GradientBoostingRegressor()
model4.fit(X_train, y_train)
print( mean_squared_error(y_test , model4.predict(X_test)) )

params = {"n_estimators": [50,100,150],
 "learning_rate": [0.1,0.05, 0.15],
 "max_depth": [2,3,4]}

gs = GridSearchCV(estimator=GradientBoostingRegressor(),
             param_grid=params,
             cv = 5
)  #model
gs.fit(X_train, y_train)

model5 = XGBRegressor()
model5.fit(X_train, y_train)
print( mean_squared_error(y_test , model5.predict(X_test)) )
```
